# Visualize the cnetral limit theorum

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
sample_size = 1000
mean_pop = 128
std_pop = 25
```

```python
ev_scores = []
```

```python
df = pd.DataFrame(columns = ["test_mean", "test_std"])
```

```python
for n_samples in range(100, 10000):
    ev_scores = []
    for i in range(n_samples):
        scores = np.random.normal(mean_pop, std_pop, sample_size)
        ev_score = np.mean(scores)
        ev_scores.append(ev_score)

    test_mean = np.mean(ev_scores)
    test_std = np.mean(ev_scores)
    df = df.append({"test_mean": test_mean, "test_std": test_std}, ignore_index = True)
```

```python
df
```

to get error value

```python
import math
sample_mean = mean_pop
sample_std = std_pop/math.sqrt(sample_size)
```

```python
error_mean_list = []
error_std_list = []

for test_mean in df["test_mean"]:
    error_mean = test_mean - sample_mean
    error_std = test_std - sample_std
    error_mean_list.append(error_mean)
    error_std_list.append(error_std)
    
df["error_mean"] = error_mean_list
df["error_std"] = error_std_list
```

```python
from matplotlib import pyplot
pyplot.plot(df["error_mean"])
```
